## 写作<Mirror>镜子
* 第一个假设，世界是有平行时空的。
 
* 第二个假设，超能力者只能观察，不能改变；而且观察的数量是有限的，每次只能观察一个。观察的过程要消耗自己的物理尺度也就是自己生命值。
  
* 第三个假设，在一个封闭的平时时空里，一个人的生命如果能跨越其物理尺度，那么他就能跃迁到另一个平行时空（且维度只能比此前高一个等级里)。
  
* 最后一个假设，平行时空最高等级与最低等级的维度之差不能大于等于10个等级.
